```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# Use raw string or double backslashes for Windows paths
metadata_path = r'D:\Euphoscan\data\metadata.csv'  # Change this to your actual absolute path if needed
image_dir = r'D:\Euphoscan\data\images'             # Make sure this folder exists and contains your images

# Read metadata CSV
df = pd.read_csv(metadata_path)
print("Dataset info:")
print(df.head())

# Plot class distribution
plt.figure(figsize=(8,5))
df['dx'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Skin Lesion Type')
plt.ylabel('Count')
plt.show()

# Function to show sample images with debug info
def show_sample_images(df, num=5):
    plt.figure(figsize=(15, 5))
    for i, (_, row) in enumerate(df.sample(num).iterrows()):
        img_path = os.path.join(image_dir, row['image_id'] + '.jpg')
        print(f"Trying to open image: {img_path}")
        if os.path.exists(img_path):
            img = Image.open(img_path)
            plt.subplot(1, num, i+1)
            plt.imshow(img)
            plt.title(row['dx'])
            plt.axis('off')
        else:
            print(f"Image NOT found: {img_path}")
    plt.show()

show_sample_images(df)

# Check missing images
missing = []
for img_id in df['image_id']:
    img_path = os.path.join(image_dir, img_id + '.jpg')
    if not os.path.exists(img_path):
        missing.append(img_id)

print(f"Total missing images: {len(missing)}")
print("Example missing image IDs:", missing[:10])

# Print current working directory and files (for your info)
print("Current working directory:", os.getcwd())
print("Files in current directory:", os.listdir())
```
