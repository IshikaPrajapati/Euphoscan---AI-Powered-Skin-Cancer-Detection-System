```python
# 02_preprocessing.ipynb

import pandas as pd
import os
from PIL import Image, ImageOps
import numpy as np
from sklearn.model_selection import train_test_split
import random

# Load metadata
metadata_path = r'D:\Euphoscan\data\metadata.csv'
df = pd.read_csv(metadata_path)
df['filename'] = df['image_id'] + '.jpg'

# Split train/val
train_df, val_df = train_test_split(df, test_size=0.2, stratify=df['dx'], random_state=42)

# Folder for images
image_dir = r'D:\Euphoscan\data\images'

# Function to load & augment images
def load_images(df, target_size=(224, 224)):
    X = []
    y = []
    for _, row in df.iterrows():
        img_path = os.path.join(image_dir, row['filename'])
        if os.path.exists(img_path):
            img = Image.open(img_path).resize(target_size)

            # Example random horizontal flip
            if random.random() < 0.5:
                img = ImageOps.mirror(img)

            img = np.array(img, dtype=np.float32) / 255.0
            X.append(img)
            y.append(row['dx'])
    return np.array(X), np.array(y)

# Load training data
X_train, y_train = load_images(train_df)
print(f"Train shape: {X_train.shape}")

# Load validation data
X_val, y_val = load_images(val_df)
print(f"Validation shape: {X_val.shape}")

# Save to .npy if you want (optional)
np.save('../data/X_train.npy', X_train)
np.save('../data/y_train.npy', y_train)
np.save('../data/X_val.npy', X_val)
np.save('../data/y_val.npy', y_val)
```
